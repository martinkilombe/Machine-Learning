```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df= pd.read_csv("D:\GIT1\Machine-Learning\Regression\housing.csv")
df
```

The regression is going to cover the Numerical(area,bedrooms,bathrooms,stories,parking) and Non-numerical features(mainroad,guestroom,basement,hotwaterheating,airconditioning,prefarea,furnishingstatus)  

```python
#Numerical Features and inputs
df_num = df[['price','area','bedrooms','bathrooms','stories','parking']]
df_num
```

```python
#fit regression
reg = linear_model.LinearRegression()
reg.fit(df_num[['area','bedrooms','bathrooms','stories','parking']],df_num.price)
```

```python
#getting coefficients and intercepts ---- y = m1*3.31115495e+02 + m2*1.67809788e+05 + m3*1.13374016e+06 + m4*5.47939810e+05 + m5*3.77596289e+05 - 145734.4894558806
reg.coef_ , reg.intercept_
```

```python
#predicting using method 1
reg.predict([[9960,3,2,2,2]])
```

```python
#predicting using method 2
# Predicting using named features
prediction_input = pd.DataFrame({
    'area': [9960],
    'bedrooms': [3],
    'bathrooms': [2],
    'stories': [2],
    'parking': [2]
})
prediction = reg.predict(prediction_input)
print(prediction)
```
