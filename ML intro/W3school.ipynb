Polynomial regression

```python
x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

mymodel = np.poly1d(np.polyfit(x, y, 3))

myline = np.linspace(1, 22, 100)

plt.scatter(x, y)
plt.plot(myline, mymodel(myline))
plt.show()
```

```python
from sklearn.metrics import r2_score

mymodel = np.poly1d(np.polyfit(x, y, 3))

print(r2_score(y, mymodel(x)))
```

```python
#predict future values
mymodel =np.poly1d(np.polyfit(x,y,3))
speed = mymodel(17)
print(speed)
```

Machine Learning - Train/Test

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(2)
```

```python
x = np.random.normal(3,1,100)
y= np.random.normal(150,40,100)/x
```

```python
plt.scatter(x,y)
plt.show()
```

```python
#Split Into Train/Test --The training set should be a random selection of 80% of the original data.--The testing set should be the remaining 20%.
#train set
train_x = x[:80]
train_y = y[:80]
```

```python
#test set
test_x = x[80:]
test_y =y[80:]
```

```python
#Display the Training Set
plt.scatter(train_x,train_y)
plt.show()
```

```python
#Display the Testing Set
plt.scatter(test_x,test_y)
plt.show()
```

```python
#fitting a polynomial model
mymodel1 = np.poly1d(np.polyfit(train_x,train_y,4))

```

```python
myline = np.linspace(0,6,100)
plt.scatter(train_x,train_y)
plt.plot(myline,mymodel1(myline))
plt.show()
```

```python
#goodness of the model --r2
from sklearn.metrics import r2_score
r2 = r2_score(train_y, mymodel1(train_x))
print(r2)
```

```python
#predicting
mymodel1(5)
```
